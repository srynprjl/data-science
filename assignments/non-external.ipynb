### Stack 

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if len(self.stack) == 0:
            print("Stack Empty")
            return None
        else:
            return self.stack.pop()

    def top(self):
        if len(self.stack) == 0:
            print("Stack Empty")
            return None
        else:
            return self.stack[-1]

    def display(self):
        for i in self.stack:
            print(i , end=" ")

stack = Stack()

stack.push(10)
stack.push(20)
stack.push(30)
stack.push(40)
stack.push(50)

stack.display()
print()
print(f"Pop: {stack.pop()}")

print(f"Top: {stack.top()}")
stack.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None

    def push(self, data):
        node = Node(data)
        node.next = self.top
        self.top = node

    def pop(self):
        if not self.top:
            print("Stack is empty")
            return None
        pop = self.top.data
        self.top = self.top.next
        return pop

    def top(self):
        if self.top:
            return self.top.data
        return None

    def display(self):
        i = self.top
        while i is not None:
            print(f"{i.data}", end=" ")
            i = i.next
        print()

stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)
stack.push(40)
stack.push(50)
stack.push(60)
stack.display()

print(stack.pop())
stack.display()
```

## Queue

```python
class Queue:
    def __init__(self, max_size=10):
        self.queue = []
        self.max = max_size

    def enqueue(self, add_item):
        if len(self.queue) >= self.max:
            print("Queue is Full")
        else:
            self.queue.append(add_item)

    def dequeue(self):
        if not self.queue:
            print("Queue Empty")
            return None
        else:
            pop_item = self.queue.pop(0)
            return pop_item

    def display(self):
        if not self.queue:
            print("Queue is empty")
        else:
            for i in self.queue:
                print(i, end= " ")
            print()

queue = Queue(max_size=5)
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.enqueue(4)
queue.enqueue(5)
queue.enqueue(6)

queue.display()

queue.dequeue()
queue.dequeue()

queue.display()

```

## Graph

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_nodes(self, node):
        if node not in self.graph:
            self.graph[node] = []

    def add_edge(self, edge1, edge2):
        self.add_nodes(edge1)
        self.add_nodes(edge2)
        self.graph[edge1].append(edge2)
        self.graph[edge2].append(edge1)

    def get_neighbours(self, node):
        return self.graph.get(node, [])

    def display(self):
        for nodes, neighbours in self.graph.items():
            print(f"{nodes} -> {neighbours} \n")

graph = Graph()
graph.add_edge('A', 'B')
graph.add_edge('A', 'C')
graph.add_edge('B', 'D')
graph.add_edge('C', 'D')
graph.add_nodes('E')
graph.display()

graph.get_neighbours('B')
```

## Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_start(self, data):
        node = Node(data)
        node.next = self.head
        self.head = node
        return self.head

    def insert_at_end(self, data):
        node = Node(data)
        if(self.head == None):
            self.head = node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = node

    def insert(self, data, index):
        if(index == 0):
            self.insert_at_start(data)
            return
        last = self.head
        for i in range(index -1):
            if last.next is None:
                print("Index is out of bounds. ")
                return
            last = last.next
        node = Node(data)
        node.next = last.next
        last.next = node

    def delete(self, data):
        temp = self.head
        if temp and temp.data == data:
            self.head == temp.next
            return
        prev = None
        while temp and temp.data != data:
            prev = temp
            temp = temp.next
        if not temp:
            return

        prev.next = temp.next

    def traverse(self):
        head = self.head
        list = f"{ head.data } "
        while head.next:
            list += f" -> {head.next.data}"
            head = head.next
        print(list)

    def get(self, index):
        temp = self.head
        for i in range(index):
            if i is not None:
                temp = temp.next
            else:
                break
        return temp.data

l = LinkedList()
l.insert_at_end(10)
l.insert_at_end(12)
l.insert_at_end(14)
l.insert_at_end(3)
l.insert_at_end(12)
l.insert_at_start(3)
l.insert(11, 2)

l.traverse()
l.delete(11)


l.traverse()
```

## Tree

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.children = []
        self.parent = None

    def add_child(self, child):
        child.parent = self
        self.children.append(child)

    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent
        return level

    def display(self):
        print(f" {"    " * self.get_level()} -- {self.data}" )
        for i in self.children:
            i.display()


a = Node(1)
b = Node(2)
c = Node(3)
a.add_child(b)
a.add_child(c)

d = Node(4)
e = Node(5)


f = Node(6)
g = Node(7)

b.add_child(d)
b.add_child(e)

c.add_child(f)
c.add_child(g)

a.display()
```
