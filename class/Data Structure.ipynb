## List
- ordered collection
- mutable ( can changed )
- Allows duplicates

```python
my_list = [1, 2, 3, 3, 4]
```

```python
x = my_list.append(6)
print(my_list) # adds the number to the end of the list.
print(x)
```

```python
x = my_list.insert(10, 2) 
print(my_list) # (x, y) , adds the number y into index x
print(x)
```

```python
x = my_list.remove(3)
print(my_list)
# no x in list, then it will give ValueError(x not in list)
# if x in list, it removes the numver
# if duplicate number in lists, it removes the first

print(x)
```

```python
x = my_list.pop()
print(my_list)
print(f"This is the popped element: {x}")
## removes the last elements and returns the value.
```

```python
x = my_list.sort()
print(my_list) # sorts the list in ascending order
print(x)
```

```python
x = my_list.reverse()
print(my_list) # reverses the list
print(x)
```

## Tuples

### Checking time difference between list & tuples

```python
import time
start = time.time()
print(start)
for _ in range(500_000):
    a = [1, 2, 3, 4]
end = time.time()
print(f"{end} \n")
print(f"List Time: {end - start}")

start = time.time()
for _ in range(500_000):
    a = (1, 2, 3, 4)
end = time.time()
print(f"Tuple Time: {end - start}")

```

```python
import sys
a = [1, 2, 3, 4, 5]
b = (1, 2, 3, 4, 5)

print(f"List: {sys.getsizeof(a)}")
print(f"Tuple: {sys.getsizeof(b)}")
```

# Set

- Unordered
- No duplicates
- Fast membership test

```python
a = {1, 2, 3, 4}
b = {3, 4, 5}

print(a.union(b)) # returns a set of all unique items in A & B
print(a.intersection(b)) # returns a set of same items in A & B
print(a.difference(b)) # is in A but not in B
print(a.symmetric_difference(b)) # returns set of all unique items removing the intersected value (union - intersection)
```

```python
import time
start = time.time()
print(start)
for _ in range(500_000):
    a = {1, 2, 3, 4}
end = time.time()
print(f"{end} \n")
print(f"List Time: {end - start}")

start = time.time()
for _ in range(500_000):
    a = [1, 2, 3, 4]
end = time.time()
print(f"Tuple Time: {end - start}")

```
