## Arithematic Operator

```python
print(1+2)
print(3-4)
print(5*6)
print(7/2)
print(7%2)
print(7//2)
```

## Comparison Operator

```python
print(1==3)
print(1!=3)
print(1>3)
print(1<3)
print(1>=3)
print(1<=3)
```

## Logical Operator

```python
print(True and False)
print(not True)
print(False or True)
```

## Assignment Operator

```python
x=5
print(x)
x+=5
print(x)
x-=5
print(x)
x*=5
print(x)
x/=5
print(x)
x%=5
print(x)
x//=5
print(x)
```

## Bitwise Operator

```python
print(3&1)
print(3|1)
print(3^1)
print(~4)
print(3<<2)
print(5>>1)
```

## Membership Operator

```python
print("s" in "string")
print("d" in "nepal")

x = [1, 2, 3, 4, 5]
print(1 in x)
print(6 in x)
```

## Identity Operator

```python
a = 5
b = 5
print(a is b)

a = [1, 2, 3]
b = [1, 2, 3]
print(a is b)
```
